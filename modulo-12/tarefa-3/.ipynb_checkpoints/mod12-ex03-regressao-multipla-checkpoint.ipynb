### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
# Importando as bibliotecas
from seaborn import load_dataset, scatterplot
from patsy import dmatrices
from statsmodels.api import OLS
from matplotlib.pyplot import axhline
from statsmodels.formula.api import ols
from numpy import log, power, mean
from pandas import read_csv
````

````python
# Carregando a base de dados
tips = load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
# 1.1
format = 'tip ~ smoker + net_bill + 1'

y, X = dmatrices(format, tips)
````

````python
# 1.2
reg = OLS(y,X).fit()

reg.summary()
````

````python
# 1.3
reg = ols(format, data=tips).fit()
tips['res'] = reg.resid

scatterplot(x='net_bill', y='res', data=tips)
axhline(y=0, color='r', linestyle='--')
````

````python
# 1.4 - Log
formatLog = 'tip ~ smoker + log(net_bill) + 1'

reg = ols(formatLog, data=tips).fit()
tips['res_log'] = reg.resid

scatterplot(x='net_bill', y='res_log', data=tips)
axhline(y=0, color='r', linestyle='--')
````

````python
# 1.4 - Polinômio
formatPoli = 'tip ~ smoker + power(net_bill, 2) + 1'

reg = ols(formatPoli, data=tips).fit()
tips['res_power'] = reg.resid

scatterplot(x='net_bill', y='res_power', data=tips)
axhline(y=0, color='r', linestyle='--')
````

Minha escolha é o modelo com log, devido o professor utilizar muito na aula. É difícil escolher um entre esses dois modelos quando não se ver grandes diferenças entre eles. Aparentemente, ambos são iguais. Ainda não sei identificar qual o melhor modelo, até o r quadrado deles são próximos, se for escolher pelo maior r quadrado, seria o modelo com polinômio. Mas é difícil escolher sem o por quê.

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, diner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
# 2.1 + 2.2
format2 = 'tip ~ + smoker + net_bill + 1'

y, X = dmatrices(format2, tips)

reg2 = OLS(y,X).fit()

reg2.summary()
````

````python
# 2.3
tips['res'] = reg2.resid

scatterplot(x='net_bill', y='res', data=tips)
axhline(y=0, color='r', linestyle='--')
````

````python
# 2.4 - Log
format24Log = 'tip ~ smoker + log(net_bill) + 1'

reg24Log = ols(format24Log, data=tips).fit()
tips['res_log'] = reg24Log.resid

scatterplot(x='net_bill', y='res_log', data=tips)
axhline(y=0, color='r', linestyle='--')

reg24Log.summary()
````

````python
# 2.4 - Polinômio

format24Poli = 'tip ~ smoker + power(net_bill, 2) + 1'

reg24Poli = ols(format24Poli, data=tips).fit()
tips['res_power'] = reg24Poli.resid

scatterplot(x='net_bill', y='res_power', data=tips)
axhline(y=0, color='r', linestyle='--')

reg24Poli.summary()
````

Escolho o modelo com polinômio, pois possui um $R^2$ maior.
