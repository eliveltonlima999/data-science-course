### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
# 1 Importando bibliotecas
from seaborn import load_dataset, regplot, scatterplot
from numpy import linspace, log, random, power, exp
from matplotlib.pyplot import plot, xlim, ylim, axhline
from statsmodels.formula.api import ols
from sklearn.metrics import r2_score
````

````python
# 1-A - Parte 1 - Carregando a base
tips = load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
# 1-A - Parte 2 - Regressão linear simples com stats models
reg = ols('tip ~ net_bill', data=tips).fit()
reg.summary()
````

````python
# 1-B Gráfico de dispersão
regplot(y='tip', x='net_bill', data=tips[tips['tip_pct']<.5])

# Gerando dados para a linha dos 10%
x = linspace(0, tips['net_bill'].max(), 100)
y = .1*x

# Plotando a linha dos 10%
plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
xlim(left=0)
ylim(bottom=0)
````

````python
# 2-A Modelo de regressão
reg = ols('tip_pct ~ net_bill', data=tips).fit()
reg.summary()
````

````python
# 2-B Gráfico de disperssão
regplot(y='tip_pct', x='net_bill', data=tips[tips['tip_pct']<.5])

# Gerando dados para a linha dos 10%
x = linspace(0, tips['net_bill'].max(), 100)
y = .1*x

# Plotando a linha dos 10%
plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
xlim(left=0)
ylim(bottom=0)
````

````python
pred_tip_pct = reg.fittedvalues
pred_tip_pct
````

````python
pred_tip = pred_tip_pct * tips['net_bill'].values
pred_tip
````

````python
# 4.3
r2_score(tips['tip'].values, pred_tip)
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
tips['tip_pct'] = log((tips['net_bill']-(tips['net_bill'].min()-.1))) + random.randn(244)*.05
tips['tip_pct'] = tips['tip_pct'] - tips['tip_pct'].mean()
regplot(x='net_bill', y='tip_pct', data=tips)
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg = ols('tip_pct ~ net_bill + power(tip_pct,2)', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

scatterplot(x='tip_pct', y='res', data=tips, alpha=.75)
axhline(y=0, color='r', linestyle='--')
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg = ols('net_bill ~ exp(tip_pct)', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

scatterplot(x='tip_pct', y='res', data=tips, alpha=.75)
axhline(y=0, color='r', linestyle='--')
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg = ols('net_bill ~ log(tip_pct+.1)', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

scatterplot(x='tip_pct', y='res', data=tips, alpha=.75)
axhline(y=0, color='r', linestyle='--')
````
