#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
from pandas import read_csv as rcsv, crosstab
from seaborn import countplot as ctt, catplot as ctt
from matplotlib import pyplot as plt

df_sinasc = rcsv('sinasc_RO_2019.csv')
df_sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df_sinasc.loc[(df_sinasc['APGAR5'] <= 3), 'APGAR_CAT'] = 'asfixia grave'
df_sinasc.loc[(df_sinasc['APGAR5'] >= 4) & (df_sinasc['APGAR5'] <= 5), 'APGAR_CAT'] = 'asfixia moderada'
df_sinasc.loc[(df_sinasc['APGAR5'] >= 6) & (df_sinasc['APGAR5'] <= 7), 'APGAR_CAT'] = 'asfixia leve'
df_sinasc.loc[(df_sinasc['APGAR5'] >= 8), 'APGAR_CAT'] = 'normal'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
ctt(
    x='GESTACAO', 
    y='APGAR5',
    hue='SEXO',
    kind='bar',
    estimator='sum',
    data=df_sinasc,
    errorbar=None,
    height=10,
)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
crosstab_apgar_gest = crosstab(df_sinasc['GESTACAO'], df_sinasc['APGAR5'])
crosstab_apgar_gest
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Somando linhas
freq_sum = crosstab_apgar_gest.sum(axis=1)
print(freq_sum)

# Dividindo cada elemento pela soma para obter os percetuais de linha somando 100
rows_percentage = crosstab_apgar_gest.div(freq_sum, axis=0)

# Verificando se está somando 1
rows_percentage.sum(axis=1)

rows_percentage
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Gerando grafico
rows_percentage.plot.bar(figsize=[15, 10])
```

Acredito que os médicos estão certos. O resultado da escala de APGAR é considerado normal quando apresenta um valor superior a 7. Isso indica um menor risco de complicações após o nascimento, em comparação com bebês que apresentam resultados inferiores. Embora resultados inferiores a 7 não indiquem que o bebê terá complicações de saúde no futuro, eles estão associados a um maior risco de doenças como paralisia cerebral e epilepsia, principalmente quando identificados no 5º minuto de vida. Que na análise é o APGAR5.
