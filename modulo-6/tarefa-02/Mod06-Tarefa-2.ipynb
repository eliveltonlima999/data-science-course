```python
from pandas import read_csv as rcsv
from seaborn import boxplot as bpt, displot as dpt
from matplotlib import pyplot as plt

df = rcsv('sinasc_RO_2019.csv')
df.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
bpt(y="IDADEMAE", data=df).set_ylabel('IDADE DA MÃE')

plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
bpt(y="PESO", data=df).set_ylabel('PESO DO BEBÊ')

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
dpt(
    data = df,
    x = 'IDADEMAE',
    kind="hist",
    bins=30, # Ajustando o número de bins para uma melhor visualização
    kde=True, # Adicionando a estimativa de densidade kernel para suavisar o gráfico
    color='skyblue', # Definindo a cor das barras
    edgecolor='black', # Definindo a cor das bordas das barras
).set_axis_labels('IDADE DA MÃE', 'FREQUÊNCIA')

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
dpt(
    data = df,
    x = 'PESO',
    kind="hist",
    bins=30, # Ajustando o número de bins para uma melhor visualização
    kde=True, # Adicionando a estimativa de densidade kernel para suavisar o gráfico
    color='green', # Definindo a cor das barras
    edgecolor='blue', # Definindo a cor das bordas das barras
).set_axis_labels('PESO DO BEBÊ', 'FREQUÊNCIA')

plt.show()
```
