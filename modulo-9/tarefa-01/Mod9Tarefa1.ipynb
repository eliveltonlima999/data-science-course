# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_1 = sinasc_raw[['munResNome', 'IDADEMAE', 'IDADEPAI']]
sinasc_1.groupby('munResNome').apply('mean')
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
# Filtrando por data '2019-07-31'
date_search = sinasc_raw[sinasc_raw['DTNASC'] == '2019-07-31']

# Peganso somente o necessario
columns_necessary = date_search[['ESCMAE', 'PESO', 'SEXO']]

# Fazendo agrupamento e tirando a media do peso do bebe
average_weight = columns_necessary.groupby(['ESCMAE', 'SEXO']).agg('mean')
average_weight
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
group = sinasc_raw[['munResNome', 'IDADEMAE', 'IDADEPAI', 'PARTO']].groupby('munResNome').agg({
    'IDADEMAE': [('Média', 'mean'), ('Máximo', 'max'), ('Mínimo', 'min')],
    'IDADEPAI': [('Média', 'mean'), ('Máximo', 'max'), ('Mínimo', 'min')],
    'munResNome': [('Total', 'count')],
    'PARTO': [('Total', 'count')],
}).sort_values(by=[('PARTO', 'Total')], ascending=False).tail(2)

group

# Castanheiras
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
import warnings
warnings.filterwarnings('ignore')

group2 = sinasc_raw[['munResNome', 'QTDFILVIVO', 'IDADEPAI', 'DTNASC', 'PARTO']]
group2['DTNASC'] = pd.to_datetime(group2['DTNASC'])
group2_marco = group2.loc[group2['DTNASC'].dt.month == 3]

group2_marco.groupby('munResNome')['PARTO'].count().sort_values(ascending=False).head(3)

city_marco = group2_marco.loc[group2_marco['munResNome'] == 'Porto Velho']
city_marco.groupby('munResNome')[['QTDFILVIVO', 'IDADEPAI']].agg([('Média', 'mean'), 'max', 'min']).round(2)

# Porto Velho
```
