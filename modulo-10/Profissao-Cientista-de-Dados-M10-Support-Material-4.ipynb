# Gráficos de perfís de médias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
ax = sns.pointplot(
    x="sex", 
    y="tip_pct",
    data=tips, 
    dodge=True, 
    errorbar=('ci', 90),
)
```

```python
ax = sns.pointplot(
    x="day", 
    y="tip_pct", 
    hue='sex',
    data=tips, 
    dodge=True, 
    errorbar=('ci', 95)
)

plt.legend(
    loc='lower center',
    bbox_to_anchor=(0.5, -.30),
    ncol=3
)
```
